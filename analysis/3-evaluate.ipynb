```python
import sys
import torch
import polars as pl
import plotnine as p9

sys.path.append('../src/')
import utils

images = torch.load('../data/images.pt')
losses = pl.read_csv('../data/losses.csv')
model = torch.load('../data/model.pt')

train = images[0:40000]
test = images[40000:50000]

plot = (
    p9.ggplot(
        losses,
        p9.aes(
            x='epoch',
            y='losses'
        )
    ) +
    p9.geom_line(
        size=3,
        color='white'
    ) +
    p9.geom_line(
        size=1
    ) +
    p9.theme_538() +
    p9.theme(
        plot_title=p9.element_text(face='bold')
    ) +
    p9.labs(
        title='MAE Loss by Epoch',
        y='MAE Loss (Log)',
        x='Epoch'
    ) +
    p9.scale_y_continuous(
        trans='log'
    )
)

plot
```
